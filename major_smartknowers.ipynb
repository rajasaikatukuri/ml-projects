```python
#Splitting Independent features and Dependent features
X=major.drop(['income_ >50K'],axis=1)
y=major['income_ >50K']
```

```python
#Importing StandardScalar
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
#Importing LogisticRegression,SVC,KNeighboursClassifier,RandomForestClassifier,DecisionTreeClasssifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# list of models
models_list = [LogisticRegression(max_iter=1000), KNeighborsClassifier(), RandomForestClassifier(random_state=0), DecisionTreeClassifier(random_state=0), SVC(kernel='linear')]
```

```python
# creating a dictionary that contains hyperparameter values for the above mentioned models


hyper_parameters = {
    

    'log_reg_hyperparameters': {
        
        'C' : [1,5,7,10,15,20]
    },


    'KNN_hyperparameters' : {
        
        'n_neighbors' : [3]
    },


    'random_forest_hyperparameters' : {
        
        'n_estimators' : [50, 100,150,200],
        'criterion' : ['gini', 'entropy']
    },
    
    'decision_tree_hyperparameters': {
        'max_features': ['sqrt', 'log2'],
        'criterion' : ['gini', 'entropy']
        
    },
    
    'svc_hyperparameters': {
        'C' : [1,3],
        'kernel' : ['linear']
    }


}
```

```python
model_keys = list(hyper_parameters.keys())
print(model_keys)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#Using train_test_spilt
x_train,x_test,Y_train,Y_test=train_test_split(X,y,stratify=y,random_state=143,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#Using StandardScalar
x_train=ss.fit_transform(x_train)
x_test=ss.fit_transform(x_test)
```

```python
print(x_train.std())
print(x_test.std())
```

```python
#fitting the model
#making predictions
#print confusion matrix
#print classification_report
#gridsearch_CV
def ModelSelection(list_of_models, hyperparameters_dict):

  result = []

  i = 0

  for model in list_of_models:

    key = model_keys[i]

    params = hyperparameters_dict[key]

    i += 1

    print(model)
    print(params)
    print('---------------------------------')


    classifier = GridSearchCV(model, params, cv=5)

    # fitting the data to classifier
    classifier.fit(x_train,Y_train)
    y_pred=classifier.predict(x_test)
    
    print(confusion_matrix(Y_test,y_pred))
    tn, fp, fn, tp=confusion_matrix(Y_test,y_pred).ravel()
    print(classification_report(Y_test,y_pred))
    print('miscalcualtion rate percentage: ',((fp+fn)/(tn+fp+fn+tp))*100)
    print('                                                                  ')
    print('******************************************************************')
    print('------------------------------------------------------------------')

    result.append({
        'model used' : model,
        'highest score' : classifier.best_score_,
        'best hyperparameters' : classifier.best_params_
    })

  result_dataframe = pd.DataFrame(result, columns = ['model used','highest score','best hyperparameters'])

  return result_dataframe
```

```python
ModelSelection(models_list, hyper_parameters)
```
