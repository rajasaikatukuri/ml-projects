```python
from datetime import datetime
ipl['date']=ipl.loc[:,'date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
```

```python
ipl_final=pd.get_dummies(ipl,columns=['venue','bat_team','bowl_team'])
```

```python
ipl_final.head()
```

```python
ipl_final.info()
```

```python
X_train=ipl_final.drop('total',axis=1)[ipl_final['date'].dt.year<=2016]
X_test=ipl_final.drop('total',axis=1)[ipl_final['date'].dt.year>=2017]
```

```python
y_train=ipl_final[ipl_final['date'].dt.year<=2016]['total'].values
y_test=ipl_final[ipl_final['date'].dt.year>=2017]['total'].values
```

```python
X_train.drop('date', axis=1, inplace=True)
X_test.drop('date', axis=1, inplace=True)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import RandomizedSearchCV
```

```python
ridge=Ridge()
param={'alpha':[1e-20,1e-15,1e-10,1e-5,1e-3,1e-1,1,2,3,5,10,15,20,30,40,50]}
ridge_reg=RandomizedSearchCV(ridge,param_distributions=param,scoring='neg_mean_squared_error',cv=7,n_iter=20)
```

```python
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.best_params_
```

```python
ridge_reg.best_score_
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso()
param={'alpha':[1e-20,1e-15,1e-10,1e-5,1e-3,1e-1,1,2,3,5,10,15,20,30,40,50]}
lasso_reg=RandomizedSearchCV(lasso,param,scoring='neg_mean_squared_error',cv=7,n_iter=20)
```

```python
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.best_params_
```

```python
lasso_reg.best_score_
```
